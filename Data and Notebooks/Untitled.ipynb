```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('housing.csv', nrows = 10000)
df.head()
```

```python
plt.scatter("housing_median_age", "median_house_value", data = df)
plt.xlabel("housing_median_age")
plt.ylabel("median_house_value")
plt.show()
```

```python
plt.scatter("longitude", "median_house_value", data = df)
plt.xlabel("longitude")
plt.ylabel("median house value")
plt.show()
```

```python
plt.scatter("population", "median_house_value", data = df)
plt.xlabel("population")
plt.ylabel("median house value")
plt.show()
```

```python
plt.scatter(df["median_income"], df["median_house_value"])
plt.xlabel("median income")
plt.ylabel("median house value")
plt.show()
```
